```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x))
ax.set_title('simple plot')
ax.set_xlabel('wave length')
ax.set_ylabel('amplitude')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
ax.set_title('simple plot')
ax.set_xlabel('wave length')
ax.set_ylabel('amplitude')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x), label = 'sin')
ax.plot(x, np.cos(x), label = 'cos')
ax.set_title('simple plot')
ax.set_xlabel('wave length')
ax.set_ylabel('amplitude')
ax.legend()
# ax.legend(loc=1)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x), label = 'sin')
ax.plot(x, np.cos(x), label = 'cos')
ax.plot(x, np.tan(x), label = 'tan')
ax.set_title('simple plot')
ax.set_xlabel('wave length')
ax.set_ylabel('amplitude')
ax.legend()
# ax.legend(loc=1)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 5, 10)
print(x)
ax.plot(x, np.sin(x), label = 'sin', color = 'red')
ax.plot(x, np.cos(x), label = 'cos', color = 'green')
ax.set_title('simple plot')
ax.set_xlabel('wave length')
ax.set_ylabel('amplitude')
ax.legend()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
print(x)
ax.plot(x, np.sin(x))
ax.set_xlim(-5, 15)
ax.set_ylim(-3, 3);
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,6))
x = np.linspace(0, 10, 1000)
axs[0,0].plot(x, np.sin(x))
axs[0,1].plot(x, np.cos(x), color = 'maroon')
axs[1,0].plot(x, np.sin(x+2), color = 'blue')
axs[1,1].plot(x, np.sin(x+4), color = 'green')
```

```python
data = {'apple': 10, 'orange': 15, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())
fig = plt.figure()
ax = plt.axes()
ax.grid(linestyle='--', color='0.85')
ax.bar(names, values);
```

```python
data = {'apple': 10, 'orange': 15, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())
fig = plt.figure()
ax = plt.axes()
#ax.grid(linestyle='--', color='0.85')
ax.bar(names, values);
ax.bar(names, values, color='teal', width = 0.35);
ax.set_xlabel('fruit')
ax.set_ylabel('quantity')
ax.set_title('fruit quantities')
```

```python
data = {'apple': 10, 'orange': 15, 'banana': 13, 'pear': 9, 'pomegranate': 11, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())
ax = plt.axes()
ax.barh(names, values);      # horizontal bar-graph
ax.set_xlabel('fruit')
ax.set_ylabel('quantity')
ax.set_title('fruit quantities');
```
